```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import random 
from fbprophet import Prophet
```

```python
data = pd.read_csv('avocado.csv')
```

```python
data.head()
```

# Exploring the Dataset

```python
data.tail()
```

```python
data_df = data.sort_values('Date')
```

```python
data_df
```

```python
plt.figure(figsize = (10,10))
plt.plot(data_df['Date'],data_df['AveragePrice'])
```

```python
plt.figure(figsize = (25,12))
sns.countplot(x = 'region',data = data)
plt.xticks(rotation = 45, fontsize = 13)
```

```python
sns.countplot(x = 'year', data = data)
```

```python
avocado_prophet_df = data_df[['Date','AveragePrice']]
```

```python
avocado_prophet_df
```

# Make Prediction


```python
avocado_prophet_df = avocado_prophet_df.rename(columns = {'Date':'ds','AveragePrice':'y'})
```

```python
avocado_prophet_df
```

```python
m = Prophet()
```

```python
m.fit(avocado_prophet_df)
```

# Forecast The Future

```python
future = m.make_future_dataframe(periods = 365)
forecast = m.predict(future)
```

```python
forecast
```

```python
figure = m.plot(forecast, xlabel = 'Date', ylabel = 'Price')
```

```python
figure = m.plot_components(forecast)
```

# Making Prediction part-2

```python
# making prediction for specific region..here i am taking west region
avocado_df_sample = data[data['region'] == 'West']
```

```python
avocado_df_sample
```

```python
avocado_df_sample = avocado_df_sample.sort_values('Date')
```

```python
avocado_df_sample
```

```python
plt.figure(figsize = (10,10))
plt.plot(avocado_df_sample['Date'],avocado_df_sample['AveragePrice'])
```

```python
avocado_df = avocado_df_sample[['Date','AveragePrice']]
avocado_d_sample = avocado_df.rename(columns = {'Date':'ds','AveragePrice':'y'})
```

```python
avocado_d_sample
```

```python
m = Prophet()
m.fit(avocado_d_sample)
```

```python
future = m.make_future_dataframe(periods = 720)
forecast = m.predict(future)
```

```python
figure = m.plot(forecast, xlabel = 'Date', ylabel = 'Price')
```

```python
figure = m.plot_components(forecast)
```
